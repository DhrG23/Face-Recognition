```python
import cv2
import numpy as np
import dlib
from PIL import Image
from io import BytesIO
from IPython.display import display, update_display
import face_recognition

# Load the known images and encode the faces
known_image_paths = ["known_person1.jpg", "known_person2.jpg"]  # Change these to your known person's images

known_encodings = []
for path in known_image_paths:
    known_image = face_recognition.load_image_file(path)
    encoding = face_recognition.face_encodings(known_image)[0]
    known_encodings.append(encoding)

# Initialize webcam capture
video_capture = cv2.VideoCapture(0)

# Display window setup for Colab
disp = display("", display_id=True)

while True:
    # Capture frame-by-frame
    ret, frame = video_capture.read()

    # Convert frame from BGR to RGB (required by face_recognition)
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Find face locations in the frame
    face_locations = face_recognition.face_locations(rgb_frame)
    face_encodings = face_recognition.face_encodings(rgb_frame, face_locations)

    # Loop through the detected faces
    for face_location, face_encoding in zip(face_locations, face_encodings):
        # Compare the encodings with known encodings
        results = face_recognition.compare_faces(known_encodings, face_encoding)

        # Draw a rectangle around the detected face
        top, right, bottom, left = face_location
        cv2.rectangle(frame, (left, top), (right, bottom), (255, 0, 0), 2)

        # Add label based on verification result
        if True in results:
            label = "Known Person"
        else:
            label = "Unknown Person"
        cv2.putText(frame, label, (left, top - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    # Convert frame from RGB to BGR for displaying with OpenCV
    bgr_frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)

    # Convert BGR frame to image for displaying in Colab
    image = Image.fromarray(bgr_frame)

    # Display the frame in Colab
    img_byte_arr = BytesIO()
    image.save(img_byte_arr, format='jpeg')
    disp.update(Image(data=img_byte_arr.getvalue()))

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release webcam capture
video_capture.release()
cv2.destroyAllWindows()
```
